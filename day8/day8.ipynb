```python
inp = open('day8.txt', 'r').read()
```

```python
from itertools import combinations
boxes = [tuple(map(int, line.split(','))) for line in inp.splitlines()]
dists = [(a, b, (sum((x - y) ** 2 for x, y in zip(a, b)))**0.5)
         for a, b in combinations(boxes, 2)]
dists.sort(key=lambda x: x[2])
C = 1000

union_find = {}
def find(x):
    if union_find[x] != x:
        union_find[x] = find(union_find[x])
    return union_find[x]
def union(x, y):
    union_find[find(x)] = find(y)
def make_set(x):
    union_find[x] = x

for box in boxes:
    make_set(box)

for a, b, dist in dists[:C]:
    union(a, b)
    
clusters = {}
for box in boxes:
    root = find(box)
    if root not in clusters:
        clusters[root] = 0
    clusters[root] += 1
sizes = sorted(clusters.values(), reverse=True)
sizes[0] * sizes[1] * sizes[2]
```
