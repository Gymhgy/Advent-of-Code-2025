```python
import numpy as np
grid = np.array([[1 if c == '@' else 0 for c in line] for line in inp.splitlines()])

from scipy.signal import convolve2d
kernel = np.array([[1,1,1],[1,0,1],[1,1,1]])
result = convolve2d(grid, kernel, mode='same', boundary='fill', fillvalue=0)
np.sum((grid == 1) & (result < 4)).item()
```

```python
import numpy as np
grid = np.array([[1 if c == '@' else 0 for c in line] for line in inp.splitlines()])

from scipy.signal import convolve2d
kernel = np.array([[1,1,1],[1,0,1],[1,1,1]])
t = 0
while True:
    result = convolve2d(grid, kernel, mode='same', boundary='fill', fillvalue=0)
    t += np.sum((grid == 1) & (result < 4)).item()
    new_grid = ((grid == 1) & (result >= 4)).astype(int)
    if np.array_equal(new_grid, grid):
        break
    grid = new_grid
t
```
