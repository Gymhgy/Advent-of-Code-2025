```python
inp = open('day5.txt', 'r').read()
```

```python
ranges, ings = inp.split('\n\n')
ranges = ranges.split('\n')
ings = list(map(int, ings.split('\n')))

ranges = [r.split('-') for r in ranges]


t = 0
for ing in ings:
    for r in ranges:
        if int(r[0]) <= ing <= int(r[1]):
            t += 1
            break

t
```

```python
ranges = sorted(ranges, key=lambda x: int(x[0]))
merged = [ranges[0]]
for current in ranges[1:]:
    last = merged[-1]
    if int(current[0]) <= int(last[1]):
        last[1] = str(max(int(last[1]), int(current[1])))
    else:
        merged.append(current)

sum(int(merged[i][1]) - int(merged[i][0]) + 1 for i in range(len(merged)))
```
