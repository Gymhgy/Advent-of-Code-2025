```python
inp = open('day7.txt', 'r').read()
```

```python
grid = inp.split('\n')
beams = {grid[0].index('S')}

s = 0
for row in grid[1:]:
    new_beams = set()
    for beam in beams:
        if row[beam] == '^':
            s += 1
            if beam + 1 < len(row):
                new_beams.add(beam + 1)
            if beam - 1 >= 0:
                new_beams.add(beam - 1)
        else :
            new_beams.add(beam)
    beams = new_beams

s
```

```python
from functools import cache

@cache
def times(r, c):
    if r == len(grid):
        return 1
    if grid[r][c] == '^':
        return times(r + 1, c - 1) + times(r + 1, c + 1)
    else:
        return times(r + 1, c)

times(0, grid[0].index('S'))
```
