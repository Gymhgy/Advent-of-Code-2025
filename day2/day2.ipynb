```python
inp = open('day2.txt', 'r').read()
```

```python
from functools import cache

@cache
def invalid(n):
    n = str(n)
    if len(n) % 2 != 0:
        return False
    half = len(n) // 2
    return n[:half] == n[half:]

t = 0
for r in inp.split(','):
    s, e = r.split('-')
    s, e = int(s), int(e)
    for i in range(s, e + 1):
        if invalid(i):
            t += i

t
```

```python
from functools import cache

@cache
def invalid(n):
    n = str(n)
    for l in range(1, len(n) // 2 + 1):
        if len(n) % l != 0:
            continue
        if n == n[:l] * (len(n) // l):
            return True
    return False

t = 0
for r in inp.split(','):
    s, e = r.split('-')
    s, e = int(s), int(e)
    for i in range(s, e + 1):
        if invalid(i):
            t += i

t
```
