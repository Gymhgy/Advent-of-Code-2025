```python
inp = open('day6.txt', 'r').read()
```

```python
all_nums = list(zip(*[map(int, line.split()) for line in inp.split('\n')[:-1]]))
ops = inp.split('\n')[-1].split()

t = 0
for nums, op in zip(all_nums, ops):
    if op == '+':
        t += sum(nums)
    elif op == '*':
        p = 1
        for n in nums:
            p *= n
        t += p

t
```

```python
all_nums = inp.split('\n')[:-1]
all_nums = [list(map(int, s.split())) for s in '\n'.join(''.join(n) for n in zip(*all_nums)).split('\n    ')]
t = 0
for nums, op in zip(all_nums, ops):
    if op == '+':
        t += sum(nums)
    elif op == '*':
        p = 1
        for n in nums:
            p *= n
        t += p

t
```
