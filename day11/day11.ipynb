```python
inp = open('day11.txt', 'r').read()
```

```python
adj = {}
for line in inp.splitlines():
    parts = line.split()
    node = parts[0][:-1]
    neighbors = parts[1:]
    adj[node] = neighbors

def num_paths(node):
    if node == 'out':
        return 1
    total = 0
    for neighbor in adj[node]:
        total += num_paths(neighbor)
    return total

num_paths('you')
```

```python
from functools import cache

@cache
def num_paths2(start, end):
    if start == end:
        return 1
    if start not in adj:
        return 0
    total = 0
    for neighbor in adj[start]:
        total += num_paths2(neighbor, end)
    return total

print(num_paths2('svr', 'fft') * num_paths2('fft', 'dac') * num_paths2('dac', 'out'))
print(num_paths2('svr', 'dac') * num_paths2('dac', 'fft') * num_paths2('fft', 'out'))
```
