```python
inp = open('day3.txt', 'r').read()
```

```python
t = 0
for line in inp.splitlines():
    nums = list(map(int, line))
    for i in range(9, 0, -1):
        if i not in nums:
            continue
        j = nums.index(i)
        if j != len(nums) - 1:
            k = max(nums[j+1:])
            t += i * 10 + k
            break

t
```

```python
t = 0

def joltage(nums, n):
    if n == 1:
        return max(nums)
    k = max(nums[:len(nums) - n + 1])
    i = nums.index(k)
    return k * 10**(n - 1) + joltage(nums[i+1:], n - 1)
    

for line in inp.splitlines():
    nums = list(map(int, line))
    t += joltage(nums, 12)

t
```
