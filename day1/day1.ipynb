```python
inp = open('day1.txt', 'r').read()
```

```python
x = 50
r = 0
for line in inp.splitlines():
    d = (line[0] == 'R') * 2 - 1
    n = int(line[1:])
    x = (x+n * d) % 100
    if x == 0:
        r += 1

r
```
